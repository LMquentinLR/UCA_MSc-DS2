## Exercise 1 (Bayesian Data Analysis, Gelman et al, Chapter 1)

Humans male have one X-chromosome and one Y-chromosome, wheras females have two X-chromosomes, each chromosome being inherited from one parent. Hemophilia is a disease that exhibits X-chromosome-linked recessive inheritance, meaning that a male who inherits the gene that causes the disease on the X-chromosome is affected, while a female carrying the gene on only one of her two X-chromosomes is not affected.

Let's consider a woman who has an affected brother and an unaffected father. This implies that her mother carries the hemophilias gene with one "good" and one "bad" gene. Let's consider the random variable $\theta$ describing the state of the woman (carrier or not carrier). 

**1) Give the prior distribution of $\theta$.**

The woman's mother carries 1 chromosome with the hemophilic gene out of 2. As the woman inherits one gene from the mother, it means that there is a chance that she will also become a carrier. We can model this probability wih a bernouill distribution with parameter $\theta$ such that:

\begin{align}
    \theta &= 1,\text{ woman is a carrier}\\
    \theta &= 0,\text{ woman is not a carrier}\\
\end{align}

The unknown quantity of interest, the state of the woman has two values. Either she's a carrier ($\theta=1$) or she's not ($\theta=0$). Based on the information provided so far, the prior distribution for $\theta$ can be expressed as a Bernouilli distribution with parameter $p=1/2$.

$$P(\theta=i)=(\frac{1}{2})^i.(\frac{1}{2})^{1-i}$$

```python
pTheta_0, pTheta_1 = 1/2, 1/2
```

**2)** We are told that the woman has two sons, neither of whom is affected. We consider the random variable $y_i = 1, 0$ which denotes if the son number i is affected or not. The outcomes of the two sons are exchangeable, and conditional on the unknown $\theta$ are indepedent. We'll denote the data $(y_1, y_2)$ as $y$.

**Relying on this information, derive the posterior probability of the woman being affected.**

\begin{align}
    P(\theta=1|Y) &= \frac{P(Y|\theta=1).P(\theta=1)}{P(Y)}\\
    &= \frac{1}{2}.\frac{P(Y|\theta=1)}{P(Y)}\\
    P(Y) &= P(Y_1=0,Y_2=0)\\
    &= P(Y_1=0, Y_2=0|\theta=0)P(\theta=0) \\
    &+ P(Y_1=0, Y_2=0|\theta=1)P(\theta=1)\\
    P(Y_1=0, Y_2=0|\theta=1)&= P(Y_1=0|\theta=1) * P(Y_2=0|\theta=1)\tag{Independence}\\
    &= \frac{1}{2} * \frac{1}{2} = 0.25\\
    P(Y_1=0, Y_2=0|\theta=0)&= P(Y_1=0|\theta=0) * P(Y_2=0|\theta=0)\tag{Independence}\\
    &= 1 * 1 = 1\\
    P(Y) &= 1 * \frac{1}{2} + 0.25 . \frac{1}{2} = 0.625  \\
\end{align}

As such:

\begin{align}
    P(Y) &= 0.625  \\
    P(\theta=1|Y) &= 0.25 \\
    &= \frac{1}{2}.\frac{P(Y|\theta=1)}{P(Y)}\\
    &= \frac{1}{2}.\frac{0.25}{0.625}\\
    &= \frac{1}{4}
\end{align}

**3)** Let's suppose that the woman has a third son who is also unaffected. 

**What is the new posterior probability $Pr(\theta=1|y_1, y_2, y_3)$ ?**

**WE CAN REUSE THE PRIOR**

\begin{align}
    P(\theta=1|Y_1, Y_2) &= 0.2\\
    P(\theta=0|Y_1, Y_2) &= 1 - P(\theta=1|Y_1, Y_2) = 0.8\\
    P(\theta=1|Y_1,Y_2, Y_3) &= \frac{P(Y_1,Y_2, Y_3, \theta=1)}{P(Y_1, Y_2, Y_3)}\\  
    &=\frac{P(Y_3|Y_1,Y_2,\theta=1).P(\theta=1|Y_1,Y_2).P(Y_1,Y_2)}{P(Y_3|Y_1, Y_2).P(Y_1, Y_2)}\\
    &=\frac{P(Y_3|\theta=1).P(\theta=1|Y_1,Y_2)}{P(Y_3)}\tag{Independence}\\
    &=\frac{P(Y_3|\theta=1).P(\theta=1|Y_1,Y_2)}{P(Y_3|\theta=0).P(\theta=0|Y_1, Y_2) + P(Y_3|\theta=1).P(\theta=1|Y_1, Y_2)}\tag{Relying on the new prior}\\
    &= \frac{0.5 * 0.2}{1*0.8 + 0.5 * 0.2}\\
    &= 0.1111...
\end{align}

## Exercise 3 (Bayesian Data Analysis, Gelman et al, Chapter 2)

An early study on *placenta previa*, a condition of pregnancy, found that on a sample of 980 births, 437 were females. We also know that the proportion of female births in the general population is of 0.485. We will denote by $\theta$ the probability of a female birth when the mother is suffering from *placenta previa*. We will assume a prior distribution p($\theta$) = Beta($\alpha$, $\beta$).

\begin{align}
n_{total}&=980\\
n_{female}&=437\\
\mu_{pop}&=0.485\\
\theta&:\text{ probability of a female birth if mother affected by PP}\\
\end{align}

**1) Write the data likelihood.**

\begin{align}
    P(Y|\theta)&=\binom{n_{total}}{n_{female}}\theta^{n_{female}}.(1-theta)^{1-n_{female}}\\
    P(Y|\theta)&= \binom{980}{437}\theta^{437}.(1-\theta)^{980-437}
\end{align}

**2) Give the posterior probability of the number of births $\theta$ (up to a constant).**
\begin{align}
    P(\theta|Y) &\propto P(Y|\theta)P(\theta) \\
    &\propto \theta^{\alpha + 437 - 1} . (1-\theta)^{\beta + 980 - 437 - 1}\\
    &\Rightarrow \theta|Y \sim Beta(\alpha+437, \beta + 980 - 437)
\end{align}
Up to a constant.

$$\theta|Y \sim Beta(\alpha+n_{female}, \beta + n_{total} - n_{female})$$

**3) How much evidence this data provide for the claim that the proportion of female births is below 0.485, the proportion of females in the general population ? You'll be summarizing information about the posterior distribution using statistics such as the median or posterior intervals.**

The first thing to observe is that the posterior is not really sensitive to the prior when alpha and beta are small. # Posterior inference (Bayesian Data Analysis, Gelman et al, Chapter 2)

This exercise illustrates how to do posterior inference using standard probability distributions introduced in the class.

Suppose you have a Beta(4,4) prior distribution on the probability $\theta$ that a coin will yield a "head" when spun in a specified manner. The coin is independently spun ten times, and "heads" appear fewer than 3 times. You don't know how many heads were seen, but only that their number is less than 3. We will denote by y the random variable giving the number of heads obtained after the 10 throws.

\begin{align}
\theta &\sim Beta(4, 4)\\
n&=10\\
K&<3\\
Y&=\text{number of head after 10 throws}\\
\end{align}
with $\theta$ a head if spun in a specific manner.

**1) Write the prior probability distribution of $\theta$ and the conditional y|$\theta$.**

Theta is the probability that a coin will yield a head.
Y is a random variable giving the number of heads over the 10 independent throws: $P(Y=y|theta)$

\begin{align}
P(\theta) &= \frac{\theta^{4 - 1}.(1-theta)^{4 - 1}}{Beta(4, 4)}\\
&\propto \theta^3.(1-\theta)^3\\
P(Y|\theta) &= \binom{n}{y}\theta^y.(1-\theta)^{n-y}\\
P(\theta|Y) &\sim Beta(y+4, n-y+4)
\end{align}

**2) Write the data likelihood.**

$$P(Y<3|\theta) = \overset{2}{\underset{i=0}{\sum}}P(Y = i|\theta) = (1-\theta)^10 + 10 \theta(1-\theta)^9 + 45\theta^2(1-\theta)^8$$

**3) Calculate the posterior density of $\theta$ (up to a constant).**

\begin{align}
P(\theta|Y<3) &\propto P(Y<3|\theta).P(\theta)\\
P(Y<3|\theta).P(\theta) &= \theta^3(1-\theta)^13 + 10 \theta^4(1-\theta)^12 + 45\theta^5(1-\theta)^11
\end{align}

**4) Plot the posterior distribution of $\theta$ (up to a constant).**

```python
theta = np.linspace(0, 1, 100)
posterior_theta = theta**3*(1-theta)**13 + 10*theta**4*(1-theta)**12 + 45*theta**5*(1-theta)**11

plt.figure()
plt.plot(theta, posterior_theta)
plt.xlabel("theat")
plt.title("Posterior distribution of theta (up to a constant)")
plt.show()
```

# Predictive prior distribution (Bayesian Data Analysis, Gelman et al, Chapter 2)

In this exercise, we show how we can incorporate all the information we have about the parameters of an experiment, in order to derive a predictive prior over the results of this experiement.

Let y be the number of 6's in 1000 independent rolls of a particular real die, which may be unfair. Let $\theta$ be the probability that the die lands on 6. We assume the following prior distribution for $\theta$:

$$
\begin{align}
Pr(\theta = \frac{1}{12}) & = 0.25 \\
Pr(\theta = \frac{1}{6}) & = 0.5 \\
Pr(\theta = \frac{1}{4}) & = 0.25.
\end{align}
$$

**1) Using the normal approximation, give the conditional distribution p(y|$\theta$).**

In order to approximate $p(y|\theta)$ by a normal distribution we need to estimate its mean and variance. We know that the experiment consists in 1000 independent rolls of the dice so we could model $p(y|\theta)$ by a binomial distribution. This distribution has a mean \mu = n * \theta where n is the number of trials and a variance $\sigma^2 = n * \theta (1-\theta)$. Therefore we can compute estimates of the mean and variance of Y for the different values of $\theta$. In the case of $\theta = 25$, for instance we would have $\mu = 1000 * 0.25 = 250$, and $\sigma^2 = 1000 * 0.25 * 0.75 = 188$. So we obtain $p(y|\theta=0.25) \sim \mathcal{N}(250, 188)$.

**2) Give an approximate prior distribution for p(y) and plot it.**

$p(y) = \sum_i p(y, \theta=p_i) = \sum_i p(y|\theta=p_i).p(\theta=p_i)$

**3) Give approximate 5%, 25%, 50%, 75%, 95% points for the distribution of y.**

$P(y) = 1/4 \mathcal{N}(83, 76) + 1/2\mathcal{N}(167, 139) + 1/4\mathcal{N}(250, 188)$

```python
y = np.arange(50, 300)
pred_prior_y = 0.25 * norm.pdf(y, 83, np.sqrt(76)) + \
                    0.5*norm.pdf(y, 167, np.sqrt(139)) + \
                    0.25*norm.pdf(y, 250, np.sqrt(188))

plt.figure()
plt.plot(pred_prior_y)
plt.show()
```
